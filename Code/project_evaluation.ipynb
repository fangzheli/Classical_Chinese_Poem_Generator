```python
Rhythm = [2,1,4,4,2,1,3,3,2,2,4,2,3,1,4,4]
Consistency = [4,1,4,3,3,1,3,3,2,2,4,2,3,1,3,3]
Fluency = [3,3,4,4,2,2,3,3,3,4,3,3,3,2,3,3]
Innovation = [3,2,5,3,1,1,4,3,5,3,5,2,2,1,3,4]
```

```python
classes = [0,1] * 8
```

```python
poem_id = [1,2,3,4] * 4
```

```python
import pandas as pd
d = {"Rhythm":Rhythm, "Consistency":Consistency, "Fluency":Fluency, "Innovation":Innovation, "classes":classes, "poem_id":poem_id}
human_eval = pd.DataFrame(data=d)
```

```python
human_eval[human_eval.classes==0].mean()
```

```python
human_eval.mean()
```

```python
import seaborn as sns
p1 = sns.catplot(x="poem_id", y="Rhythm", data=human_eval).set(title="score for Rhythm vs different poem").savefig("Rhythm.jpg")
```

```python
sns.catplot(x="poem_id", y="Consistency", data=human_eval).set(title="score for Consistency vs different poem").savefig("Consistency.jpg")
```

```python
sns.catplot(x="poem_id", y="Fluency", data=human_eval).set(title="score for Fluency vs different poem").savefig("Fluency.jpg")
```

```python
sns.catplot(x="poem_id", y="Innovation", data=human_eval).set(title="score for Innovation vs different poem").savefig("Innovation.jpg")
```
